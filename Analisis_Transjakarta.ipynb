###### `corridorID` dan `corridorName`

Untuk `corridorID` yang tidak ada `corridorName` dan sebaliknya, pengisian dilakukan dengan memanfaatkan pemetaan antara kedua kolom tersebut yang sudah tersedia.

```python
corridorNameToID = df.dropna(subset=['corridorID', 'corridorName']).drop_duplicates('corridorName').set_index('corridorName')['corridorID'].to_dict()
corridorIDtoName = df.dropna(subset=['corridorID', 'corridorName']).drop_duplicates('corridorID').set_index('corridorID')['corridorName'].to_dict()


df['corridorID'] = df.apply(lambda row: corridorNameToID.get(row['corridorName'], row['corridorID']) if pd.isnull(row['corridorID']) else row['corridorID'], axis=1)

df['corridorName'] = df.apply(lambda row: corridorIDtoName.get(row['corridorID'], row['corridorName']) if pd.isnull(row['corridorName']) else row['corridorName'], axis=1)

df[['corridorID', 'corridorName']].isnull().sum()
```

Setelah proses imputasi pada kolom corridorID dan corridorName, masih terdapat 1.125 missing value pada masing-masing kolom. Untuk mengatasinya, saya mengisi nilai yang belum terisi tersebut dengan label 'TBD', dengan asumsi bahwa data tersebut merupakan bagian dari penyesuaian koridor baru yang dilakukan oleh TransJakarta . Hal ini dilakukan karena tidak ada kolom lain yang dapat dijadikan referensi untuk pemetaan atau pengisian kedua kolom tersebut.

```python
df[['corridorID', 'corridorName']] = df[['corridorID', 'corridorName']].fillna('TBD')
```

###### `tapOutStops` dan `tapOutStopsName`

Untuk `tapOutStops` dan `TapOutStopsName`, saya mencoba mengisinya berdasarkan pemetaan yang tersedia.

```python
tapOutStopsToNameMapping = df.dropna(subset=['tapOutStops', 'tapOutStopsName']).drop_duplicates('tapOutStops').set_index('tapOutStops')['tapOutStopsName'].to_dict()

tapOutStopsNameToIDMapping = df.dropna(subset=['tapOutStops', 'tapOutStopsName']).drop_duplicates('tapOutStopsName').set_index('tapOutStopsName')['tapOutStops'].to_dict()

df['tapOutStopsName'] = df.apply(
    lambda row: tapOutStopsToNameMapping[row['tapOutStops']] if pd.isnull(row['tapOutStopsName']) and row['tapOutStops'] in tapOutStopsToNameMapping else row['tapOutStopsName'],
    axis=1
)

df['tapOutStops'] = df.apply(
    lambda row: tapOutStopsNameToIDMapping[row['tapOutStopsName']] if pd.isnull(row['tapOutStops']) and row['tapOutStopsName'] in tapOutStopsNameToIDMapping else row['tapOutStops'],
    axis=1
)
```

```python
print("Sisa missing value yang ada: ")
df[['tapOutStops', 'tapOutStopsName']].isna().sum()
```

Terdapat jumlah missing value yang cukup besar pada kolom `tapOutStops` (1.363) dan `tapOutStopsName` (1.344). Kedua kolom ini biasanya saling terkait, sehingga jika salah satu kosong, kemungkinan besar yang lain juga kosong. Penyebab utama missing value kemungkinan karena penumpang tidak melakukan tap out, masalah teknis alat, atau data memang tidak tersedia. Dari nilai unik yang ada, nama halte yang tertera merupakan nama jalan, gang, atau kelurahan yang merupakan halte-halte kecil. Untuk mengisi missing value tersebut saya memberikan keterangan 'Unknown ID' dengan asumsi halte-halte tersebut merupakan halte kecil dan baru sehingga belum atau bahkan tidak memiliki ID.

```python
df.loc[(df['tapOutStops'].isna()) & (df['tapOutStopsName'].notna()), 'tapOutStopsName'].unique()
```

```python
condition = df['tapOutStops'].isna() & df['tapOutStopsName'].notna()
df.loc[condition, 'tapOutStops'] = "Unknown ID"
```

Untuk mengisi nilai yang hilang ini, saya melakukan identifikasi lebih lanjut terhadap jenis layanan berdasarkan `corridorID` terkait. Berdasarkan informasi dari situs TransJakarta, saya mengelompokkan bahwa `corridorID` seperti 'M7B', 'JAK.48B', '3H', 'JAK.48A', dan 'M1H' termasuk layanan BRT, Non BRT, atau Mikrotrans dengan tarif reguler sebesar 3.500. Perbandingan Tingkat Ketidakpatuhan Tap Out antar Halte (Contoh: Terminal Senen vs Kampung Rambutan)

```python
from scipy.stats import chi2_contingency

halte_compare = df[df['tapOutStopsName'].isin(['Term. Senen', 'Term. Kampung Rambutan'])]
halte_compare['patuh'] = halte_compare['tapOutTime'].notnull().map({True: 'Patuh', False: 'Tidak Patuh'})
ct = pd.crosstab(halte_compare['tapOutStopsName'], halte_compare['patuh'])

chi2, p, dof, expected = chi2_contingency(ct)
print('Tabel Kontingensi:\n', ct)
print(f'Chi-square: {chi2:.2f}, p-value: {p:.4f}')

if p < 0.05:
    print('Terdapat perbedaan signifikan tingkat kepatuhan tap out antara Terminal Senen dan Kampung Rambutan.')
else:
    print('Tidak terdapat perbedaan signifikan tingkat kepatuhan tap out antara kedua terminal.')
```

* Berdasarkan data tersebut, dapat disimpulkan bahwa kelompok usia berpotensi menjadi faktor yang memengaruhi tingkat kepatuhan tap out. Pengguna dari kelompok usia yang lebih tua cenderung lebih patuh, pemahaman yang lebih baik terhadap aturan, atau kesadaran yang lebih tinggi terhadap konsekuensi jika tidak mematuhi. Sebaliknya, pengguna yang lebih muda di Terminal Senen tampak lebih rentan untuk tidak patuh, mungkin karena mereka sering terburu-buru, kurang memahami pentingnya tap out.

```python
from scipy.stats import f_oneway

# Filter data Terminal Senen
senen = df[df['tapOutStopsName'] == 'Term. Senen'].copy()
senen['patuh'] = senen['tapOutTime'].notnull().map({True: 'Patuh', False: 'Tidak Patuh'})

# Ambil usia berdasarkan kepatuhan
usia_patuh = senen[senen['patuh'] == 'Patuh']['age']
usia_tidak_patuh = senen[senen['patuh'] == 'Tidak Patuh']['age']

# Uji ANOVA
anova_stat, p_value = f_oneway(usia_patuh, usia_tidak_patuh)
print(f'Hasil uji ANOVA: F = {anova_stat:.2f}, p-value = {p_value:.4f}')

if p_value < 0.05:
    print('Terdapat perbedaan signifikan rata-rata usia antara pengguna yang patuh dan tidak patuh tap out di Terminal Senen.')
else:
    print('Tidak terdapat perbedaan signifikan rata-rata usia antara pengguna yang patuh dan tidak patuh tap out di Terminal Senen.')
```

* Namun berdasarkan hasil uji ANOVA tidak terdapat perbedaan signifikan rata-rata usia antara pengguna yang patuh dan tidak patuh tap out di Terminal Senen.
